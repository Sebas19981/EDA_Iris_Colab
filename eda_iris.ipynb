Importar **librerías**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Cargar el dataset Iris**

```python
df = sns.load_dataset("iris")
```

```python
print(df.head())        # Ver primeras filas
print(df.info())        # Info de columnas y tipos
print(df.describe())    # Estadísticas
```

**Verificar valores faltantes**

```python
print(df.isnull().sum())  # Ver cuántos valores nulos hay por columna
```

**Histogramas**

```python
df.hist(figsize=(10, 6), bins=15, color='lightblue')
plt.suptitle("Histogramas de Variables Numéricas", fontsize=16)
plt.tight_layout()
plt.show()
```

**Mapa de calor (Heatmap)**

```python
# Seleccionar solo columnas numéricas
numeric_df = df.select_dtypes(include=["float64", "int64"])

# Crear mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor de Correlaciones")
plt.show()

```

**Detectar patrones (gráfico de dispersión)**

```python
sns.pairplot(df, hue="species")
plt.suptitle("Distribución de Variables por Especie", fontsize=16, y=1.02)
plt.show()
```
